```python
import matplotlib.pyplot as plt

# verileri ayır
duygular = list(frekans.keys())
degerler = list(frekans.values())

# grafik
plt.figure(figsize=(12,6))
plt.bar(duygular, degerler)
plt.xticks(rotation=90)

plt.title("Duygu Frekans Grafiği")
plt.xlabel("Duygular")
plt.ylabel("Frekans")

plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    degerler,
    labels=duygular,
    autopct='%1.1f%%'
)

plt.title("Duygu Dağılımı")

plt.show()
```

```python
import numpy as np

data = np.array(degerler).reshape(1,-1)

plt.figure(figsize=(12,2))
plt.imshow(data)

plt.yticks([])
plt.xticks(range(len(duygular)), duygular, rotation=90)

plt.title("Duygu Heatmap")

plt.colorbar()

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(duygular, degerler, marker='o')

plt.xticks(rotation=90)
plt.title("Duygu Dağılım Grafiği")
plt.xlabel("Duygular")
plt.ylabel("Değer")

plt.show()
```
