```python
#buyuk-kucuk donusumu
text_lower = text["text_veri"].astype(str).apply(
    lambda x: " ".join(word.lower() for word in x.split())
)
text_lower.head(6)
```

```python
#noktalama işaretleri
text_punc = text_lower.str.replace('[^\w\s]','')
text_punc.head(26)
```

```python
#sayısal mesajların temizlenmesi
text_digit= text_punc.str.replace('\d','')
text_digit.head(26)
```

```python
#stopwords-bir dilde çok sık kullanılan ama metnin anlamına fazla katkı sağlamayan kelimeleri çıkaralım
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
sw = set(stopwords.words('turkish'))
text_stopwords = text_digit.apply(lambda x: " ".join(x for x in x.split() if x not in sw))
text_stopwords.head(20)
```

```python
#temizlenmiş veriyi csv dosyasına yazalım
df_out = pd.DataFrame({"temiz_veri": text_stopwords})
df_out.to_csv("temizlenmis_veri.csv", index=False, encoding="cp1254")
```
